```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import numpy as np
```

```python
url  = "https://www.ebay.com.au/sch/i.html?_from=R40&_nkw=apple+watch&_sacat=0&LH_Auction=1&_sop=1"
```

```python
# Get the HTML from the URL
r = requests.get(url)
html = r.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
items = soup.find_all('div', class_="s-item__wrapper clearfix")
```

```python
list_items = []
for item in items[1:]:
    title = item.find('div', class_="s-item__title").text
    condition = item.find('span', class_="SECONDARY_INFO").text
    price = item.find('span', class_="s-item__price").text
    timeleft = item.find('span', class_="s-item__time-left").text
    shippingcost = item.find('span', class_="s-item__logisticsCost").text
    # print(title, condition, price, timeleft, shippingcost)
    if condition == "Parts only":
        continue
    dict_item = {'title': title, 'condition': condition, 'price': price, 'timeleft': timeleft, 'shippingcost': shippingcost}
    list_items.append(dict_item)
```
